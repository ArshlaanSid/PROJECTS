# Login to your Instagram Handle
Submit with sample username and password

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
```

```python

def login(username,password):
    try:
        driver.get('https://www.instagram.com/')
    #     time.sleep(5)
        wait = WebDriverWait(driver,10)
        searchh = wait.until(EC.presence_of_element_located((By.NAME, 
                                          'username')))
        driver.find_element(By.NAME,'username').send_keys(username)
        pswrd_btn = driver.find_element(By.NAME,'password')
        pswrd_btn.send_keys(password)
        pswrd_btn.submit()
        print('login successful')
    except:
        print('something went wrong!')
```

# Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def like_unlike(no_of_posts):
    try:
        
        driver.refresh()
        for i in range(no_of_posts):
            # find post path
            WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,"//div[@class = '_aagw']")))
            posts_links = driver.find_elements(By.XPATH,"//div[@class = '_aagw']")

            posts_links[i].click()

            WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,"//span/button[@class='_abl-']/div/span")))
            
            # like_status is the button status if already liked or not
            like_status = driver.find_element(By.XPATH,'//span/button[@class="_abl-"]/div/span/*')

            like_status = like_status.get_attribute('aria-label')

            like_button = driver.find_element(By.XPATH,'//span/button[@class="_abl-"]/div/span')
            like_button.click()
            
            if like_status == 'like':
                print('you '+like_status+'d post',i+1)
            else:
                print('you '+like_status+'d post',i+1)

            close_btn = driver.find_element(By.XPATH,"//div[@class = 'x78zum5 x6s0dn4 xl56j7k xdt5ytf']")
            close_btn.click()
    except:
        print('something went wrong!')
```

# Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
If You have already seen the story.

Or The user has no story.

Or View the story if not yet seen.

```python
def story(handle_name):
    
    try:

        search_profile(handle_name)

        story_btn = driver.find_element(By.XPATH,'//div[starts-with(@class,"_aarf")]')

        #check is story is present using aria-disabled attribute

        if story_btn.get_attribute("aria-disabled") == "true":
            print("Coding Ninjas has no story")
        #     print(1)

        else:
        #     print(2)
            #if story is present check if you have viewed the story by checking height attribute 
            story_check = driver.find_element(By.XPATH,'//div[starts-with(@class,"_aarf")]/canvas')
            if (story_check.get_attribute("height") == "249") or story_check.get_attribute("height") == "78":
                print("I have already seen the story")


            #if not viewed then view the story
            else:
                viewstory = driver.find_element(By.XPATH,'//div[starts-with(@class,"_aarf")]/span')
                viewstory.click()
                time.sleep(4)
        #         print(3)
                while True:
                    try:
                        next_story = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
                        next_story = driver.find_element(By.XPATH,'//button[@aria-label="Next"]')
                        next_story.click()
                        time.sleep(4)
                    except:
        #                 print(4)
                        break
                print('you have just seen the story')
    except:
        print('something went wrong!')
```
